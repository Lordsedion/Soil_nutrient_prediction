```python
X = []
Y = []
```

```python
def trainerNN(pp_list):
    
    for i in range(len(pp_list)):
        lat = []
        
        lat.append(pp_list[i][-2])
        lat.append(pp_list[i][-1])
        X.append(lat)
        Y.append(pp_list[i][2])
```

```python
trainerNN(data.iloc[:].values)
```

```python
X = np.array(X, dtype=float)
Y = np.array(Y, dtype=float)
X, Y
```

```python
X_train, X_test, Y_train, Y_test = np.array(X[0:int(round(len(X))*0.80)]), X[int(round(len(X))*0.80):], Y[:int(round(len(X))*0.80)], Y[int(round(len(X))*0.80):]
```

```python
lr = KNeighborsRegressor(n_neighbors=4)
```

```python
lr.fit(X_train, Y_train)
```
